# thay ? bằng NaN

```python
df2 = df.replace("?",np.nan,inplace=False)
print(df2)
```

# dropna cột price

```python
df2.dropna(axis=0, subset=['price'] ,inplace=True)
print(df2)
```

# ép kiểu cho compression-ratio

```python
df2['compression-ratio'] = pd.to_numeric(df2['compression-ratio'], errors='coerce')
```

```python
mean = df2['compression-ratio'].mean()
```

```python
print(mean)
```

# replace mean trong df2

```python
df2['compression-ratio'] = df2['compression-ratio'].fillna(mean)
```

```python
print(df2)
```
